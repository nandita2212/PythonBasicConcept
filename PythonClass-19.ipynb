```python
#polymorphism with class methods:

class A():
    def capital(self):
        print("Delhi")
        
    
    def language(self):
        print("Hindi is most widely spoken language")
        
class B():
    def capital(self):
        print("Berlin")
        
    def language(self):
        print("English is used")
```

```python
obj_A = A()
obj_B = B()

for i in (obj_A, obj_B):
    i.capital()
    i.language()
```

```python
#Polymorphism with Inheritance:

# Methods in the child class that have the same name as the methods in the parent class.
```

```python
class Bird:
    def flight(self):
        print("most of the birds can fly")
        

class Ostrich(Bird):
    def flight(self):
        print("ostrichs cannot fly")
        
class sparrow(Bird):
    def flight(self):
        print("sparrow can fly")
    
    

    
obj_bird = Bird()
obj_ost = Ostrich()
obj_spr = sparrow()
```

```python
obj_bird.flight()
```

```python
obj_ost.flight()
```

```python
obj_spr.flight()
```

```python
#Polymorphis with a function and objects:

class A():
    def capital(self):
        print("Delhi")
        
    
    def language(self):
        print("Hindi is most widely spoken language")
        
class B():
    def capital(self):
        print("Berlin")
        
    def language(self):
        print("English is used")
        
def fun(obj):
    obj.capital()
    obj.language()
    
obj_A = A()
obj_B = B()

fun(obj_A)
```

```python
fun(obj_B)
```

```python
# #Abstraction in Python :

# It is used to hide the internal functionality of the function.

# The users only interact with basic implementation and inner working hidden.

# "What function does" that we can know but "how it works" that we dont know.
```

```python
# #Syntax:

# from abc import ABC
# class ClassName(ABC):
```

```python
from abc import ABC, abstractmethod

class Car(ABC):
    
    @abstractmethod
    def mileage(self):
        pass
    
class Tesla(Car):
    def mileage(self):
        print("the mileage is 30kmph")
        
class Suzuki(Car):
    def mileage(self):
        print("the mileage is 25kmph")
        

t = Tesla()
t.mileage()
    
s = Suzuki()
s.mileage()

c = Car()
c.mileage()
```

```python
# # Points to remember :

# 1. An abstract class can contain the both method i.e. normal and abstract method.
# 2. It cannot be instantiated, we cannot create objects for abstract class.
```
